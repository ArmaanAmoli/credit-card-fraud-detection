```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
training = pd.read_parquet('../encodedData/encoaded_training_data_full.parquet')
training
```

```python
training_y = pd.read_parquet('../encodedData/Y_train.parquet')['is_fraud']
training_y = training_y.reset_index(drop=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
model = make_pipeline(StandardScaler(), 
                      LogisticRegression(solver='saga', max_iter=2000, random_state=42))
```

```python
training.shape
```

```python
model.fit(training,training_y)
```

```python
import joblib

filename = 'random_forest_model.joblib'

# Save the model to a file
joblib.dump(model, filename)
```
